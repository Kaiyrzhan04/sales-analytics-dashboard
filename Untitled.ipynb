```python
"""
Шаг 1 — EDA (Разведочный анализ) + Очистка данных
Online Retail II Dataset

Запуск:
    pip install pandas matplotlib seaborn
    python 01_eda_and_clean_local.py
"""

import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import seaborn as sns
import os
import warnings
warnings.filterwarnings('ignore')

# ── ПУТИ — меняй только здесь ──────────────────────────────────────
BASE_DIR    = r"C:\Users\erlan\OneDrive\Desktop\pet_project\Новая папка"
FILE_PATH   = os.path.join(BASE_DIR, "online_retail_II.csv")
OUTPUT_DIR  = os.path.join(BASE_DIR, "output")
OUTPUT_CSV  = os.path.join(OUTPUT_DIR, "clean_retail.csv")
OUTPUT_PNG  = os.path.join(OUTPUT_DIR, "01_eda_overview.png")

os.makedirs(OUTPUT_DIR, exist_ok=True)

# ── Настройки графиков ──────────────────────────────────────────────
sns.set_theme(style="whitegrid", palette="muted")
plt.rcParams['figure.figsize'] = (12, 5)
plt.rcParams['axes.titlesize'] = 14
plt.rcParams['axes.titleweight'] = 'bold'


# ═══════════════════════════════════════════════════════════════════
# 1. ЗАГРУЗКА
# ═══════════════════════════════════════════════════════════════════
print("=" * 55)
print("  1. ЗАГРУЗКА ДАННЫХ")
print("=" * 55)

df = pd.read_csv(FILE_PATH, parse_dates=['InvoiceDate'])
print(f"  Строк: {len(df):,}")
print(f"  Колонок: {df.shape[1]}")
print(f"\n  Колонки: {list(df.columns)}")
print(f"\n  Период: {df['InvoiceDate'].min().date()} — {df['InvoiceDate'].max().date()}")
print(f"\n  Типы данных:\n{df.dtypes}")


# ═══════════════════════════════════════════════════════════════════
# 2. ДИАГНОСТИКА ПРОБЛЕМ
# ═══════════════════════════════════════════════════════════════════
print("\n" + "=" * 55)
print("  2. ДИАГНОСТИКА ПРОБЛЕМ")
print("=" * 55)

print(f"\n  Пропуски:")
nulls = df.isnull().sum()
for col, n in nulls[nulls > 0].items():
    print(f"    {col}: {n:,} ({n/len(df)*100:.1f}%)")

print(f"\n  Дубликаты: {df.duplicated().sum():,}")

cancellations = df['Invoice'].astype(str).str.startswith('C').sum()
print(f"  Отмены (Invoice начинается с 'C'): {cancellations:,}")

neg_qty   = (df['Quantity'] < 0).sum()
neg_price = (df['Price'] <= 0).sum()
print(f"  Отрицательное количество: {neg_qty:,}")
print(f"  Нулевая/отрицательная цена: {neg_price:,}")


# ═══════════════════════════════════════════════════════════════════
# 3. ОЧИСТКА
# ═══════════════════════════════════════════════════════════════════
print("\n" + "=" * 55)
print("  3. ОЧИСТКА ДАННЫХ")
print("=" * 55)

df_clean    = df.copy()
start_count = len(df_clean)

df_clean = df_clean.drop_duplicates()
print(f"  После удаления дублей:     {len(df_clean):>10,} (убрано {start_count - len(df_clean):,})")

df_clean = df_clean[~df_clean['Invoice'].astype(str).str.startswith('C')]
print(f"  После удаления отмен:      {len(df_clean):>10,}")

df_clean = df_clean[df_clean['Quantity'] > 0]
df_clean = df_clean[df_clean['Price'] > 0]
print(f"  После фильтра qty/price:   {len(df_clean):>10,}")

# Добавляем полезные колонки
df_clean['HasCustomer'] = df_clean['Customer ID'].notna()
df_clean['Revenue']     = df_clean['Quantity'] * df_clean['Price']
df_clean['Year']        = df_clean['InvoiceDate'].dt.year
df_clean['Month']       = df_clean['InvoiceDate'].dt.month
df_clean['MonthName']   = df_clean['InvoiceDate'].dt.strftime('%b')
df_clean['DayOfWeek']   = df_clean['InvoiceDate'].dt.day_name()
df_clean['Hour']        = df_clean['InvoiceDate'].dt.hour
df_clean['YearMonth']   = df_clean['InvoiceDate'].dt.to_period('M')

df_clean = df_clean.rename(columns={'Customer ID': 'CustomerID'})

print(f"\n  Итого чистых строк: {len(df_clean):,}")
print(f"  Убрано всего: {start_count - len(df_clean):,} ({(start_count - len(df_clean))/start_count*100:.1f}%)")


# ═══════════════════════════════════════════════════════════════════
# 4. БАЗОВАЯ СТАТИСТИКА
# ═══════════════════════════════════════════════════════════════════
print("\n" + "=" * 55)
print("  4. БАЗОВАЯ СТАТИСТИКА")
print("=" * 55)

print(f"\n  💰 Выручка:             £{df_clean['Revenue'].sum():>12,.0f}")
print(f"  📦 Уникальных заказов:  {df_clean['Invoice'].nunique():>12,}")
print(f"  👥 Уникальных клиентов: {df_clean['CustomerID'].nunique():>12,}")
print(f"  🛍️  Уникальных товаров:  {df_clean['StockCode'].nunique():>12,}")
avg_order = df_clean.groupby('Invoice')['Revenue'].sum().mean()
print(f"  📊 Средний чек:         £{avg_order:>12,.2f}")


# ═══════════════════════════════════════════════════════════════════
# 5. ВИЗУАЛИЗАЦИИ
# ═══════════════════════════════════════════════════════════════════
print("\n" + "=" * 55)
print("  5. СТРОИМ ГРАФИКИ...")
print("=" * 55)

fig, axes = plt.subplots(2, 2, figsize=(16, 10))
fig.suptitle('Online Retail II — EDA Overview', fontsize=16, fontweight='bold', y=1.01)

# График 1: Выручка по месяцам
ax1 = axes[0, 0]
monthly = df_clean.groupby('YearMonth')['Revenue'].sum().reset_index()
monthly['YearMonth'] = monthly['YearMonth'].astype(str)
ax1.bar(monthly['YearMonth'], monthly['Revenue'], color='steelblue', alpha=0.85)
ax1.set_title('Выручка по месяцам')
ax1.set_ylabel('Выручка (£)')
ax1.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'£{x/1000:.0f}k'))
ax1.tick_params(axis='x', rotation=45)

# График 2: Топ-10 стран (без UK)
ax2 = axes[0, 1]
top_countries = (df_clean[df_clean['Country'] != 'United Kingdom']
                 .groupby('Country')['Revenue'].sum()
                 .sort_values(ascending=True).tail(10))
top_countries.plot(kind='barh', ax=ax2, color='coral', alpha=0.85)
ax2.set_title('Топ-10 стран (без UK)')
ax2.set_xlabel('Выручка (£)')
ax2.xaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'£{x/1000:.0f}k'))

# График 3: Продажи по дням недели
ax3 = axes[1, 0]
day_order = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
day_rev = df_clean.groupby('DayOfWeek')['Revenue'].sum().reindex(day_order)
ax3.bar(day_rev.index, day_rev.values, color='mediumseagreen', alpha=0.85)
ax3.set_title('Выручка по дням недели')
ax3.set_ylabel('Выручка (£)')
ax3.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'£{x/1000:.0f}k'))
ax3.tick_params(axis='x', rotation=30)

# График 4: Топ-10 товаров
ax4 = axes[1, 1]
top_products = (df_clean.groupby('Description')['Revenue']
                .sum().sort_values(ascending=True).tail(10))
top_products.plot(kind='barh', ax=ax4, color='mediumpurple', alpha=0.85)
ax4.set_title('Топ-10 товаров по выручке')
ax4.set_xlabel('Выручка (£)')
ax4.xaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'£{x/1000:.0f}k'))

plt.tight_layout()
plt.savefig(OUTPUT_PNG, dpi=150, bbox_inches='tight')
plt.show()
print(f"  ✅ График сохранён: {OUTPUT_PNG}")


# ═══════════════════════════════════════════════════════════════════
# 6. СОХРАНЯЕМ ЧИСТЫЕ ДАННЫЕ
# ═══════════════════════════════════════════════════════════════════
df_clean.drop(columns=['YearMonth']).to_csv(OUTPUT_CSV, index=False)
print(f"  ✅ Чистые данные: {OUTPUT_CSV}")
print(f"     {len(df_clean):,} строк × {df_clean.shape[1]} колонок")
print("\n" + "=" * 55)
print("  ГОТОВО!")
print("=" * 55)
```

```python
"""
Шаг 2 — RFM-анализ клиентов
Online Retail II Dataset

Запуск:
    python 02_rfm_analysis_local.py
"""

import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import seaborn as sns
import os
import warnings
warnings.filterwarnings('ignore')

# ── ПУТИ ───────────────────────────────────────────────────────────
BASE_DIR   = r"C:\Users\erlan\OneDrive\Desktop\pet_project\Новая папка"
INPUT_CSV  = os.path.join(BASE_DIR, "output", "clean_retail.csv")
OUTPUT_DIR = os.path.join(BASE_DIR, "output")
os.makedirs(OUTPUT_DIR, exist_ok=True)

sns.set_theme(style="whitegrid", palette="muted")


# ═══════════════════════════════════════════════════════════════════
# 1. ЗАГРУЗКА ЧИСТЫХ ДАННЫХ
# ═══════════════════════════════════════════════════════════════════
print("=" * 55)
print("  1. ЗАГРУЗКА ДАННЫХ")
print("=" * 55)

df = pd.read_csv(INPUT_CSV, parse_dates=['InvoiceDate'])

# Для RFM нужны только клиенты с CustomerID
df_rfm = df[df['CustomerID'].notna()].copy()
df_rfm['CustomerID'] = df_rfm['CustomerID'].astype(int)

print(f"  Всего строк:    {len(df):,}")
print(f"  Строк с CustomerID: {len(df_rfm):,}")
print(f"  Уникальных клиентов: {df_rfm['CustomerID'].nunique():,}")


# ═══════════════════════════════════════════════════════════════════
# 2. СЧИТАЕМ RFM-МЕТРИКИ
# ═══════════════════════════════════════════════════════════════════
print("\n" + "=" * 55)
print("  2. СЧИТАЕМ RFM-МЕТРИКИ")
print("=" * 55)

# Дата отсчёта — день после последней покупки в датасете
snapshot_date = df_rfm['InvoiceDate'].max() + pd.Timedelta(days=1)
print(f"  Дата отсчёта (snapshot): {snapshot_date.date()}")

rfm = df_rfm.groupby('CustomerID').agg(
    LastPurchase = ('InvoiceDate', 'max'),
    Frequency    = ('Invoice', 'nunique'),
    Monetary     = ('Revenue', 'sum')
).reset_index()

rfm['Recency'] = (snapshot_date - rfm['LastPurchase']).dt.days
rfm = rfm.drop(columns=['LastPurchase'])

print(f"\n  Статистика RFM:")
print(rfm[['Recency','Frequency','Monetary']].describe().round(1).to_string())


# ═══════════════════════════════════════════════════════════════════
# 3. СКОРИНГ 1–4
# ═══════════════════════════════════════════════════════════════════
print("\n" + "=" * 55)
print("  3. СКОРИНГ (квантили 1–4)")
print("=" * 55)

# Recency: меньше = лучше (4 = самый свежий)
rfm['R_score'] = pd.qcut(rfm['Recency'],   q=4, labels=[4,3,2,1]).astype(int)
# Frequency и Monetary: больше = лучше
rfm['F_score'] = pd.qcut(rfm['Frequency'].rank(method='first'), q=4, labels=[1,2,3,4]).astype(int)
rfm['M_score'] = pd.qcut(rfm['Monetary'].rank(method='first'),  q=4, labels=[1,2,3,4]).astype(int)

rfm['RFM_Score'] = rfm['R_score'].astype(str) + rfm['F_score'].astype(str) + rfm['M_score'].astype(str)
rfm['RFM_Total'] = rfm['R_score'] + rfm['F_score'] + rfm['M_score']


# ═══════════════════════════════════════════════════════════════════
# 4. СЕГМЕНТАЦИЯ
# ═══════════════════════════════════════════════════════════════════
print("\n" + "=" * 55)
print("  4. СЕГМЕНТАЦИЯ КЛИЕНТОВ")
print("=" * 55)

def segment(row):
    r, f, m = row['R_score'], row['F_score'], row['M_score']
    if r >= 4 and f >= 4 and m >= 4:
        return 'Champions'
    elif r >= 3 and f >= 3:
        return 'Loyal Customers'
    elif r >= 4 and f <= 2:
        return 'Recent Customers'
    elif r >= 3 and f <= 2:
        return 'Promising'
    elif r <= 2 and f >= 3 and m >= 3:
        return 'At Risk'
    elif r <= 2 and f >= 4:
        return 'Cant Lose Them'
    elif r == 1 and f == 1:
        return 'Lost'
    else:
        return 'Need Attention'

rfm['Segment'] = rfm.apply(segment, axis=1)

seg_stats = rfm.groupby('Segment').agg(
    Клиентов  = ('CustomerID', 'count'),
    Avg_R     = ('Recency',   'mean'),
    Avg_F     = ('Frequency', 'mean'),
    Avg_M     = ('Monetary',  'mean'),
    Total_Rev = ('Monetary',  'sum')
).round(1).sort_values('Total_Rev', ascending=False)

print(f"\n{seg_stats.to_string()}")


# ═══════════════════════════════════════════════════════════════════
# 5. ВИЗУАЛИЗАЦИИ
# ═══════════════════════════════════════════════════════════════════
print("\n" + "=" * 55)
print("  5. СТРОИМ ГРАФИКИ...")
print("=" * 55)

fig, axes = plt.subplots(2, 2, figsize=(16, 11))
fig.suptitle('RFM-анализ клиентов', fontsize=16, fontweight='bold', y=1.01)

COLORS = {
    'Champions':       '#2ecc71',
    'Loyal Customers': '#27ae60',
    'Recent Customers':'#3498db',
    'Promising':       '#1abc9c',
    'At Risk':         '#e67e22',
    'Cant Lose Them':  '#e74c3c',
    'Need Attention':  '#f39c12',
    'Lost':            '#95a5a6',
}

# График 1: Клиенты по сегментам
ax1 = axes[0, 0]
seg_count = rfm['Segment'].value_counts()
colors = [COLORS.get(s, '#bdc3c7') for s in seg_count.index]
bars = ax1.barh(seg_count.index, seg_count.values, color=colors, alpha=0.9)
ax1.set_title('Количество клиентов по сегментам')
ax1.set_xlabel('Клиентов')
for bar, val in zip(bars, seg_count.values):
    ax1.text(bar.get_width() + 10, bar.get_y() + bar.get_height()/2,
             f'{val:,}', va='center', fontsize=9)

# График 2: Выручка по сегментам
ax2 = axes[0, 1]
seg_rev = rfm.groupby('Segment')['Monetary'].sum().sort_values(ascending=True)
colors2 = [COLORS.get(s, '#bdc3c7') for s in seg_rev.index]
seg_rev.plot(kind='barh', ax=ax2, color=colors2, alpha=0.9)
ax2.set_title('Выручка по сегментам')
ax2.set_xlabel('Выручка (£)')
ax2.xaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'£{x/1000:.0f}k'))

# График 3: Scatter R vs M, размер = Frequency
ax3 = axes[1, 0]
scatter_data = rfm.sample(min(1500, len(rfm)), random_state=42)
seg_list = scatter_data['Segment'].unique()
palette = [COLORS.get(s, '#bdc3c7') for s in seg_list]
for seg, color in zip(seg_list, palette):
    subset = scatter_data[scatter_data['Segment'] == seg]
    ax3.scatter(subset['Recency'], subset['Monetary'],
                s=subset['Frequency']*3, alpha=0.5, label=seg, color=color)
ax3.set_title('Recency vs Monetary (размер = Frequency)')
ax3.set_xlabel('Recency (дней)')
ax3.set_ylabel('Monetary (£)')
ax3.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'£{x/1000:.0f}k'))
ax3.legend(fontsize=7, loc='upper right')

# График 4: Heatmap средних значений RFM по сегментам
ax4 = axes[1, 1]
heatmap_data = rfm.groupby('Segment')[['R_score','F_score','M_score']].mean().round(2)
heatmap_data.columns = ['R score', 'F score', 'M score']
sns.heatmap(heatmap_data, annot=True, fmt='.2f', cmap='RdYlGn',
            ax=ax4, linewidths=0.5, vmin=1, vmax=4)
ax4.set_title('Средние RFM-оценки по сегментам')

plt.tight_layout()
out_png = os.path.join(OUTPUT_DIR, "02_rfm_analysis.png")
plt.savefig(out_png, dpi=150, bbox_inches='tight')
plt.show()
print(f"  ✅ График сохранён: {out_png}")


# ═══════════════════════════════════════════════════════════════════
# 6. СОХРАНЯЕМ RFM-ТАБЛИЦУ
# ═══════════════════════════════════════════════════════════════════
out_rfm = os.path.join(OUTPUT_DIR, "rfm_segments.csv")
rfm.to_csv(out_rfm, index=False)
print(f"  ✅ RFM-таблица сохранена: {out_rfm}")

print("\n" + "=" * 55)
print("  ИТОГ ПО СЕГМЕНТАМ:")
print("=" * 55)
for seg in rfm['Segment'].value_counts().index:
    n   = (rfm['Segment'] == seg).sum()
    rev = rfm[rfm['Segment'] == seg]['Monetary'].sum()
    pct = n / len(rfm) * 100
    print(f"  {seg:<20} {n:>5} клиентов ({pct:4.1f}%)  |  £{rev:>10,.0f}")

print("\n  Следующий шаг: загрузка в SQL → 03_load_to_sql.py")
```

```python
"""
Шаг 3 — Загрузка данных в SQLite + аналитические запросы
Online Retail II Dataset

Запуск:
    python 03_load_to_sql.py
"""

import pandas as pd
import sqlite3
import os

# ── ПУТИ ───────────────────────────────────────────────────────────
BASE_DIR    = r"C:\Users\erlan\OneDrive\Desktop\pet_project\Новая папка"
OUTPUT_DIR  = os.path.join(BASE_DIR, "output")
CLEAN_CSV   = os.path.join(OUTPUT_DIR, "clean_retail.csv")
RFM_CSV     = os.path.join(OUTPUT_DIR, "rfm_segments.csv")
DB_PATH     = os.path.join(OUTPUT_DIR, "sales.db")


# ═══════════════════════════════════════════════════════════════════
# 1. ЗАГРУЗКА CSV
# ═══════════════════════════════════════════════════════════════════
print("=" * 55)
print("  1. ЗАГРУЗКА CSV-ФАЙЛОВ")
print("=" * 55)

df_sales = pd.read_csv(CLEAN_CSV, parse_dates=['InvoiceDate'])
df_rfm   = pd.read_csv(RFM_CSV)

print(f"  sales:   {len(df_sales):,} строк")
print(f"  rfm:     {len(df_rfm):,} строк")


# ═══════════════════════════════════════════════════════════════════
# 2. СОЗДАЁМ SQLite БАЗУ И ТАБЛИЦЫ
# ═══════════════════════════════════════════════════════════════════
print("\n" + "=" * 55)
print("  2. СОЗДАЁМ БАЗУ ДАННЫХ")
print("=" * 55)

conn = sqlite3.connect(DB_PATH)
cur  = conn.cursor()

# Удаляем старые таблицы если есть
cur.executescript("""
    DROP TABLE IF EXISTS sales;
    DROP TABLE IF EXISTS rfm;
    DROP TABLE IF EXISTS monthly_revenue;
    DROP TABLE IF EXISTS country_revenue;
    DROP TABLE IF EXISTS top_products;
    DROP TABLE IF EXISTS segment_summary;
""")

# Таблица sales
cur.execute("""
CREATE TABLE sales (
    Invoice     TEXT,
    StockCode   TEXT,
    Description TEXT,
    Quantity    INTEGER,
    InvoiceDate TEXT,
    Price       REAL,
    CustomerID  REAL,
    Country     TEXT,
    HasCustomer INTEGER,
    Revenue     REAL,
    Year        INTEGER,
    Month       INTEGER,
    MonthName   TEXT,
    DayOfWeek   TEXT,
    Hour        INTEGER
)
""")

# Таблица rfm
cur.execute("""
CREATE TABLE rfm (
    CustomerID  INTEGER,
    Recency     INTEGER,
    Frequency   INTEGER,
    Monetary    REAL,
    R_score     INTEGER,
    F_score     INTEGER,
    M_score     INTEGER,
    RFM_Score   TEXT,
    RFM_Total   INTEGER,
    Segment     TEXT
)
""")

conn.commit()
print(f"  ✅ Таблицы созданы: sales, rfm")


# ═══════════════════════════════════════════════════════════════════
# 3. ЗАГРУЖАЕМ ДАННЫЕ
# ═══════════════════════════════════════════════════════════════════
print("\n" + "=" * 55)
print("  3. ЗАГРУЖАЕМ ДАННЫЕ В БД")
print("=" * 55)

# Приводим дату к строке для SQLite
df_sales['InvoiceDate'] = df_sales['InvoiceDate'].astype(str)

df_sales.to_sql('sales', conn, if_exists='replace', index=False)
df_rfm.to_sql('rfm',   conn, if_exists='replace', index=False)
conn.commit()

print(f"  ✅ sales: {cur.execute('SELECT COUNT(*) FROM sales').fetchone()[0]:,} строк")
print(f"  ✅ rfm:   {cur.execute('SELECT COUNT(*) FROM rfm').fetchone()[0]:,} строк")


# ═══════════════════════════════════════════════════════════════════
# 4. АНАЛИТИЧЕСКИЕ ЗАПРОСЫ
# ═══════════════════════════════════════════════════════════════════
print("\n" + "=" * 55)
print("  4. АНАЛИТИЧЕСКИЕ ЗАПРОСЫ")
print("=" * 55)

# ── Запрос 1: Выручка по месяцам ───────────────────────────────────
print("\n  📅 Выручка по месяцам:")
q1 = """
    SELECT
        Year,
        Month,
        MonthName,
        ROUND(SUM(Revenue), 2)      AS Total_Revenue,
        COUNT(DISTINCT Invoice)     AS Orders,
        COUNT(DISTINCT CustomerID)  AS Customers,
        ROUND(SUM(Revenue) / COUNT(DISTINCT Invoice), 2) AS Avg_Order_Value
    FROM sales
    GROUP BY Year, Month
    ORDER BY Year, Month
"""
df_q1 = pd.read_sql(q1, conn)
df_q1.to_sql('monthly_revenue', conn, if_exists='replace', index=False)
print(df_q1.to_string(index=False))


# ── Запрос 2: Топ-10 стран по выручке ─────────────────────────────
print("\n\n  🌍 Топ-10 стран по выручке:")
q2 = """
    SELECT
        Country,
        ROUND(SUM(Revenue), 2)      AS Total_Revenue,
        COUNT(DISTINCT Invoice)     AS Orders,
        COUNT(DISTINCT CustomerID)  AS Customers,
        ROUND(AVG(Revenue), 2)      AS Avg_Line_Revenue
    FROM sales
    GROUP BY Country
    ORDER BY Total_Revenue DESC
    LIMIT 10
"""
df_q2 = pd.read_sql(q2, conn)
df_q2.to_sql('country_revenue', conn, if_exists='replace', index=False)
print(df_q2.to_string(index=False))


# ── Запрос 3: Топ-10 товаров ──────────────────────────────────────
print("\n\n  🛍️  Топ-10 товаров по выручке:")
q3 = """
    SELECT
        StockCode,
        Description,
        SUM(Quantity)               AS Total_Qty,
        ROUND(SUM(Revenue), 2)      AS Total_Revenue,
        COUNT(DISTINCT Invoice)     AS Times_Ordered
    FROM sales
    WHERE Description IS NOT NULL
    GROUP BY StockCode, Description
    ORDER BY Total_Revenue DESC
    LIMIT 10
"""
df_q3 = pd.read_sql(q3, conn)
df_q3.to_sql('top_products', conn, if_exists='replace', index=False)
print(df_q3.to_string(index=False))


# ── Запрос 4: Сводка по RFM-сегментам ─────────────────────────────
print("\n\n  👥 Сводка по RFM-сегментам:")
q4 = """
    SELECT
        r.Segment,
        COUNT(*)                        AS Customers,
        ROUND(AVG(r.Recency), 1)        AS Avg_Recency_Days,
        ROUND(AVG(r.Frequency), 1)      AS Avg_Orders,
        ROUND(AVG(r.Monetary), 2)       AS Avg_Revenue,
        ROUND(SUM(r.Monetary), 2)       AS Total_Revenue
    FROM rfm r
    GROUP BY r.Segment
    ORDER BY Total_Revenue DESC
"""
df_q4 = pd.read_sql(q4, conn)
df_q4.to_sql('segment_summary', conn, if_exists='replace', index=False)
print(df_q4.to_string(index=False))


# ── Запрос 5: Champions — детальный список ─────────────────────────
print("\n\n  🏆 Champions (топ-15 по выручке):")
q5 = """
    SELECT
        r.CustomerID,
        r.Recency,
        r.Frequency,
        ROUND(r.Monetary, 2) AS Monetary,
        r.RFM_Score,
        r.Segment,
        s.Country
    FROM rfm r
    JOIN (
        SELECT CustomerID, Country
        FROM sales
        WHERE CustomerID IS NOT NULL
        GROUP BY CustomerID
    ) s ON r.CustomerID = s.CustomerID
    WHERE r.Segment = 'Champions'
    ORDER BY r.Monetary DESC
    LIMIT 15
"""
df_q5 = pd.read_sql(q5, conn)
print(df_q5.to_string(index=False))


# ═══════════════════════════════════════════════════════════════════
# 5. ИТОГ
# ═══════════════════════════════════════════════════════════════════
print("\n" + "=" * 55)
print("  5. ИТОГ")
print("=" * 55)

tables = cur.execute("SELECT name FROM sqlite_master WHERE type='table'").fetchall()
print(f"\n  Таблицы в БД ({DB_PATH}):")
for t in tables:
    count = cur.execute(f"SELECT COUNT(*) FROM {t[0]}").fetchone()[0]
    print(f"    📋 {t[0]:<25} {count:>8,} строк")

conn.close()

print(f"""
  ✅ База данных готова: sales.db
  
  Следующие шаги:
    → Power BI: подключись к sales.db через Get Data → SQLite
    → Или продолжаем: 04_forecast.py (прогноз продаж)
""")
print("=" * 55)
```

```python
"""
Шаг 4 — Прогноз продаж (ИСПРАВЛЕННАЯ ВЕРСИЯ)
Online Retail II Dataset

Запуск:
    python 04_forecast.py
"""

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import matplotlib.dates as mdates
import sqlite3
import os
import warnings
warnings.filterwarnings('ignore')

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# ── ПУТИ ───────────────────────────────────────────────────────────
BASE_DIR   = r"C:\Users\erlan\OneDrive\Desktop\pet_project\Новая папка"
OUTPUT_DIR = os.path.join(BASE_DIR, "output")
DB_PATH    = os.path.join(OUTPUT_DIR, "sales.db")

# ═══════════════════════════════════════════════════════════════════
# 1. ЗАГРУЗКА
# ═══════════════════════════════════════════════════════════════════
print("=" * 55)
print("  1. ЗАГРУЗКА ДАННЫХ ИЗ БД")
print("=" * 55)

conn = sqlite3.connect(DB_PATH)
df = pd.read_sql("SELECT * FROM monthly_revenue ORDER BY Year, Month", conn)
conn.close()

df['Date'] = pd.to_datetime(df['Year'].astype(str) + '-' + df['Month'].astype(str) + '-01')
df = df.sort_values('Date').reset_index(drop=True)
df['t'] = np.arange(len(df))

print(f"  Месяцев: {len(df)}")
print(f"  Период:  {df['Date'].min().strftime('%b %Y')} — {df['Date'].max().strftime('%b %Y')}")

# ═══════════════════════════════════════════════════════════════════
# 2. ПРИЗНАКИ
# ═══════════════════════════════════════════════════════════════════
df['sin_month'] = np.sin(2 * np.pi * df['Month'] / 12)
df['cos_month'] = np.cos(2 * np.pi * df['Month'] / 12)
df['is_q4']     = df['Month'].isin([10, 11, 12]).astype(int)

features = ['t', 'sin_month', 'cos_month', 'is_q4']
X = df[features].values
y = df['Total_Revenue'].values

# ═══════════════════════════════════════════════════════════════════
# 3. ОБУЧЕНИЕ
# ═══════════════════════════════════════════════════════════════════
print("\n" + "=" * 55)
print("  2. ОБУЧЕНИЕ МОДЕЛИ")
print("=" * 55)

TRAIN_SIZE = len(df) - 3
X_train, X_test = X[:TRAIN_SIZE], X[TRAIN_SIZE:]
y_train, y_test = y[:TRAIN_SIZE], y[TRAIN_SIZE:]

model = LinearRegression()
model.fit(X_train, y_train)

y_pred_train = model.predict(X_train)
y_pred_test  = model.predict(X_test)

mae  = mean_absolute_error(y_test, y_pred_test)
rmse = np.sqrt(mean_squared_error(y_test, y_pred_test))
r2   = r2_score(y_train, y_pred_train)
mape = np.mean(np.abs((y_test - y_pred_test) / y_test)) * 100

print(f"  R²   (train): {r2:.4f}")
print(f"  MAE  (test):  £{mae:,.0f}")
print(f"  RMSE (test):  £{rmse:,.0f}")
print(f"  MAPE (test):  {mape:.1f}%")

# ═══════════════════════════════════════════════════════════════════
# 4. ПРОГНОЗ НА 3 МЕСЯЦА
# ═══════════════════════════════════════════════════════════════════
print("\n" + "=" * 55)
print("  3. ПРОГНОЗ НА 3 МЕСЯЦА ВПЕРЁД")
print("=" * 55)

last_date    = df['Date'].max()
future_dates = pd.date_range(last_date + pd.offsets.MonthBegin(1), periods=3, freq='MS')
future_month = future_dates.month
future_t     = np.arange(len(df), len(df) + 3)

future_X = np.column_stack([
    future_t,
    np.sin(2 * np.pi * future_month / 12),
    np.cos(2 * np.pi * future_month / 12),
    (future_month.isin([10, 11, 12])).astype(int),
])

future_pred = model.predict(future_X).clip(min=0)

future_df = pd.DataFrame({
    'Date':     future_dates,
    'Forecast': future_pred,
    'Month':    future_month,
})

print(f"\n  {'Месяц':<15} {'Прогноз':>15}")
print(f"  {'-'*32}")
for _, row in future_df.iterrows():
    print(f"  {row['Date'].strftime('%B %Y'):<15} £{row['Forecast']:>12,.0f}")
print(f"\n  Итого 3 месяца: £{future_pred.sum():,.0f}")

# ═══════════════════════════════════════════════════════════════════
# 5. ГРАФИКИ (ИСПРАВЛЕННЫЕ)
# ═══════════════════════════════════════════════════════════════════
print("\n" + "=" * 55)
print("  4. СТРОИМ ГРАФИКИ...")
print("=" * 55)

fig, axes = plt.subplots(2, 2, figsize=(16, 10))
fig.suptitle('Прогноз продаж — Linear Regression', fontsize=16, fontweight='bold', y=1.01)

# ── График 1: Факт vs Прогноз (весь период) ──────────────────────
ax1 = axes[0, 0]
ax1.plot(df['Date'].values, y, 'o-', color='steelblue',
         lw=2, ms=4, label='Факт')
ax1.plot(df['Date'].values[:TRAIN_SIZE], y_pred_train, '--',
         color='orange', lw=1.5, alpha=0.8, label='Модель (train)')
ax1.plot(df['Date'].values[TRAIN_SIZE:], y_pred_test, 's--',
         color='tomato', lw=2, ms=6, label='Модель (test)')
ax1.plot(future_df['Date'].values, future_df['Forecast'].values, 'D-',
         color='green', lw=2, ms=8, label='Прогноз')
ax1.axvline(df['Date'].iloc[TRAIN_SIZE], color='gray', ls=':', alpha=0.6)
ax1.set_title('Факт vs Прогноз (весь период)')
ax1.set_ylabel('Выручка (£)')
ax1.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'£{x/1000:.0f}k'))
ax1.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
ax1.legend(fontsize=8)
plt.setp(ax1.xaxis.get_majorticklabels(), rotation=30)

# ── График 2: Zoom — последние 6 мес + прогноз ───────────────────
ax2 = axes[0, 1]
# Берём последние 6 точек через числовые индексы — без tail()
zoom_start = max(0, TRAIN_SIZE - 3)
zoom_dates_fact = df['Date'].values[zoom_start:]
zoom_y_fact     = y[zoom_start:]
test_dates      = df['Date'].values[TRAIN_SIZE:]

ax2.plot(zoom_dates_fact, zoom_y_fact, 'o-', color='steelblue', lw=2, ms=6, label='Факт')
ax2.plot(test_dates, y_pred_test, 's--', color='tomato', lw=2, ms=7, label='Модель (test)')
ax2.plot(future_df['Date'].values, future_df['Forecast'].values, 'D-',
         color='green', lw=2.5, ms=9, label='Прогноз')
for _, row in future_df.iterrows():
    ax2.annotate(f"£{row['Forecast']/1000:.0f}k",
                 xy=(row['Date'], row['Forecast']),
                 xytext=(0, 14), textcoords='offset points',
                 ha='center', fontsize=9, color='darkgreen', fontweight='bold')
ax2.set_title('Zoom: последние 6 мес + прогноз')
ax2.set_ylabel('Выручка (£)')
ax2.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'£{x/1000:.0f}k'))
ax2.xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))
ax2.legend(fontsize=9)
plt.setp(ax2.xaxis.get_majorticklabels(), rotation=30)

# ── График 3: Остатки (Residuals) ────────────────────────────────
ax3 = axes[1, 0]
residuals  = y_train - y_pred_train
bar_colors = ['tomato' if r < 0 else 'steelblue' for r in residuals]
ax3.bar(range(len(residuals)), residuals, color=bar_colors, alpha=0.75)
ax3.axhline(0, color='black', lw=1)
ax3.set_title('Остатки модели (Residuals)')
ax3.set_xlabel('Месяц (порядковый номер)')
ax3.set_ylabel('Ошибка (£)')
ax3.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'£{x/1000:.0f}k'))

# ── График 4: Сезонность ─────────────────────────────────────────
ax4 = axes[1, 1]
seasonal     = df.groupby('Month')['Total_Revenue'].mean()
month_labels = ['Jan','Feb','Mar','Apr','May','Jun',
                'Jul','Aug','Sep','Oct','Nov','Dec']
bar_colors4  = ['#e74c3c' if m in [10,11,12] else 'steelblue' for m in range(1,13)]
ax4.bar(month_labels,
        [seasonal.get(m, 0) for m in range(1, 13)],
        color=bar_colors4, alpha=0.85)
ax4.set_title('Сезонность: средняя выручка по месяцам\n(красный = Q4)')
ax4.set_ylabel('Выручка (£)')
ax4.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'£{x/1000:.0f}k'))

plt.tight_layout()
out_png = os.path.join(OUTPUT_DIR, "04_forecast.png")
plt.savefig(out_png, dpi=150, bbox_inches='tight')
plt.show()
print(f"  ✅ График сохранён: {out_png}")

# ═══════════════════════════════════════════════════════════════════
# 6. СОХРАНЯЕМ В БД
# ═══════════════════════════════════════════════════════════════════
conn = sqlite3.connect(DB_PATH)

actuals_db = df[['Date','Year','Month','MonthName','Total_Revenue']].copy()
actuals_db['Date'] = actuals_db['Date'].astype(str)
actuals_db['Type'] = 'Actual'

forecast_db = pd.DataFrame({
    'Date':          future_df['Date'].astype(str),
    'Year':          future_df['Date'].dt.year,
    'Month':         future_df['Month'],
    'MonthName':     future_df['Date'].dt.strftime('%b'),
    'Total_Revenue': future_df['Forecast'],
    'Type':          'Forecast',
})

combined = pd.concat([actuals_db, forecast_db], ignore_index=True)
combined.to_sql('revenue_with_forecast', conn, if_exists='replace', index=False)
conn.commit()
conn.close()

print(f"  ✅ Таблица revenue_with_forecast сохранена в БД")
print(f"""
{"=" * 55}
  ГОТОВО!
  R²   = {r2:.3f}
  MAPE = {mape:.1f}%
  Прогноз на 3 месяца записан в sales.db
  Следующий шаг → Power BI 🎯
{"=" * 55}
""")
```
